```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from keras.preprocessing import image

np.random.seed(0)
tf.random.set_seed(0)

import warnings
warnings.filterwarnings("ignore")
```

```python
model1 = keras.models.load_model('../input/trial-tomato/model_final_leaf (1).h5')
```

```python
li=['Tomato___Bacterial_spot', 'Tomato___Early_blight', 'Tomato___Late_blight', 'Tomato___Leaf_Mold', 'Tomato___Septoria_leaf_spot', 
 'Tomato___Spider_mites Two-spotted_spider_mite', 'Tomato___Target_Spot', 'Tomato___Tomato_Yellow_Leaf_Curl_Virus', 
 'Tomato___Tomato_mosaic_virus', 'Tomato___healthy']
```

```python
for i in range(2,5):
    path = "../input/trial101/test"
    image_path=path+str(i)+".jpg"
    new_img = image.load_img(image_path, target_size=(512,512))
    img = image.img_to_array(new_img)
    img = np.expand_dims(img, axis=0)
    img = img/255

    prediction = model1.predict(img)
    d = prediction.flatten()
    j = d.max()
    for index,item in enumerate(d):
        if item == j:
            class_name = li[index]
    print("Following is our prediction:"+class_name)    
    plt.figure(figsize = (4,4))
    plt.imshow(new_img)
    plt.axis('off')
    plt.title(class_name)
    plt.show()
```
